```python
!pip install tensorflow_addons
import cv2
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import load_model
from tensorflow_addons import metrics as mt
import tensorflow_addons as tfa
import tensorflow as tf
import numpy as np

age_model = load_model('Model1.h5')
emotion_model = load_model('model_face_emotion_efficientnetB3.h5', compile=False)

age_results={0:'6-20',1:'25-30', 2:'42-48', 3:'60-98'}
emotion_results = ['angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised']

GR = {0:(0,0,255), 1:(0,0,255), 2:(0,0,225), 3:(0,0,255)}
# Load the cascades
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")


# image = load_img(img_path, target_size=(224, 224))
# image = img_to_array(image)
# image = tf.expand_dims(image, 0)

# y_pred = emotion_model.predict(image)
# pred = tf.nn.softmax(y_pred)
# pred = np.argmax(y_pred)
# print(pred)
# print(emotion_results[dic])

# Open the video capture device
capture = cv2.VideoCapture(0)

while True:
    # Read a frame from the capture device
    ret, frame = capture.read()

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 3, 0, (30, 30))

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        face_img = frame[y:y+h, x:x+w]
        rerect_sized=cv2.resize(face_img,(224,224))
        normalized=rerect_sized/255.0
        reshaped=np.reshape(normalized,(1,224,224,3))
        reshaped = np.vstack([reshaped])
        age_result=age_model.predict(reshaped)
        
        # image = load_img(frame, target_size=(224, 224))
        image = img_to_array(rerect_sized)
        image = tf.expand_dims(image, 0)
        emotion_result = emotion_model.predict(image)
        # print(emotion_result)

        age_label=np.argmax(age_result,axis=1)[0]
        
        age_prob = (list(age_result))
        ag_prob = round(((age_prob[0][age_label])*100), 2)
        print(ag_prob)
        
        emotion_label=np.argmax(emotion_result,axis=1)[0]
        emo_prob = (list(emotion_result))
        em_prob = round(((emo_prob[0][emotion_label])*100), 2)
        print(em_prob)
      
        cv2.rectangle(frame,(x,y),(x+w,y+h),GR[age_label],2)
        cv2.rectangle(frame,(x,y-60),(x+w,y),GR[age_label],-1)
        cv2.putText(frame, f"{age_results[age_label]}:{ag_prob}", (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        cv2.putText(frame, f"{emotion_results[emotion_label]}:{em_prob}", (x, y-40),cv2.FONT_HERSHEY_SIMPLEX,0.9,(255,255,255),2)

    # Display the resulting frame
    cv2.imshow("Frame", frame)

    # Check if the user pressed 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release the capture device and destroy the window
capture.release()
cv2.destroyAllWindows()
```
